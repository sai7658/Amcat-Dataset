```python

specialization_salary = df.groupby('Specialization')['Salary'].mean().sort_values(ascending=False)

# Create a bar plot to visualize the distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=specialization_salary.index, y=specialization_salary.values)
plt.xticks(rotation=90)
plt.xlabel('Specialization')
plt.ylabel('Average Salary')
plt.title('Average Salary Distribution by Specialization')
plt.show()

```

```python

top_10_specializations = df.groupby('Specialization')['Salary'].mean().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_specializations.values, y=top_10_specializations.index)
plt.title('Top 10 Specializations with Highest Average Salary')
plt.xlabel('Average Salary')
plt.ylabel('Specialization')
plt.show()
```

* Polymer Technology has the highest average salary.
* It helps to understand that individuals with these specializations may be in higher demand or have more specialized skills
* It may indicate the potential career growth and earning potential within a company for individuals with specific specializations.
* This information can be valuable for students choosing a major, professionals seeking career advancement, or the company itself when making hiring decisions and salary adjustments.

```python

top_10_specializations_by_gender = df.groupby(['Gender', 'Specialization'])['Salary'].mean().unstack()
top_10_specializations_by_gender = top_10_specializations_by_gender.apply(lambda x: x.nlargest(10), axis=1)

plt.figure(figsize=(12, 6))
top_10_specializations_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Specializations with Highest Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.legend(title='Specialization',loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

* The specialization with the highest average salary is Polymer Technology.
* Males generally receive a higher average salary than females across most specializations.


```python
# Stacked Bar Plot for Gender and Specialization
gender_specialization = pd.crosstab(df['Gender'], df['Specialization'])
gender_specialization.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution by Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Specialization',loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

* Male Dominance: Males dominate most specializations, indicating a skewed gender distribution.

* Potential Gender Bias: The higher number of males in certain fields may reflect societal expectations and historical trends.

* Limited Female Representation: Some specializations have few female candidates, highlighting a gap in representation.


```python


programming_roles = df[df['Designation'].str.contains('Software Engineer|Programming Analyst', case=False, na=False)]
plt.figure(figsize=(10, 6))
sns.boxplot(x='Designation', y='Salary', data=programming_roles)
plt.title('Salary Distribution for Programming-Related Roles')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **Key Findings and Insights:**

**1. Gender Imbalance:** The dataset reveals a significant gender imbalance, with a majority of employees being male. This disparity is evident across various specializations, indicating a potential need for initiatives to attract and retain female talent in the tech industry.

**2. Specialization and Salary:** Certain specializations, notably Polymer Technology and related fields, tend to have higher average salaries. This insight can guide students in choosing specializations aligned with career growth and earning potential.

**3. College Tier and Performance:**  The analysis indicated that most employees graduated from Tier 2 colleges, and there appears to be a weak positive correlation between college GPA and salary.

**4. Location and Designation:** Bangalore emerges as a major hub for employees in the dataset.  The 'Software Engineer' designation is the most prevalent role, indicating the demand for software development skills in the industry.


**5. Age Distribution:** The majority of the employees fall within the 32-35 age range.

**Surprising Trends:**

* The weak positive correlation between college GPA and salary might be surprising. While higher GPAs are generally expected to be associated with higher earning potential, other factors such as experience, soft skills, and networking play a significant role in determining salary.
*  The unexpected finding that Polymer Technology had the highest average salary might be an indication of the specialized skills required in the field.
